```python
import pandas as pd
```

```python
dataFrame = pd.read_csv('/home/mapper/Documents/Python/GeoPythonExamples/BasicPythonPractice/Kumpula-June-2016-w-metadata.txt'\
                        , skiprows=8)
```

```python
# create a new column based on calculations between different variables
# default value sets the column type (ie float)

dataFrame['DIFF'] = 0.0
dataFrame['DIFF'] = dataFrame['MAX'] - dataFrame['MIN']

# you can do calculations directly without first creating the column

dataFrame['DIFF_Min'] = dataFrame['TEMP'] - dataFrame['MIN']

# convert the fahrenheit temperatures in temp column into celsius 

dataFrame['TEMP_Celsius'] = (dataFrame['TEMP'] - 32) / (9/5)

print(dataFrame)
```

```python
# filtering and updating data

w_temps = dataFrame.ix[dataFrame['TEMP_Celsius'] > 15]
print(w_temps)
```

```python
w_temps2 = dataFrame.ix[(dataFrame['TEMP_Celsius'] > 15) & (dataFrame['YEARMODA'] >= 20160615)]
print(w_temps2)
```

```python
w_temps2 = w_temps2.reset_index(drop=True)
print(w_temps2)
```

```python
# dealing with NaNs
w_temps2.loc[:4, 'TEMP_Celsius'] = None
print(w_temps2)
```

```python
# remove the NoData values
w_temps_clean = w_temps2.dropna(subset=['TEMP_Celsius'])
print(w_temps_clean)
```

```python
# fill NoData with 0
w_temps_na_filled = w_temps2.fillna(0)
print(w_temps_na_filled)
```

```python
# sort values in ascending order
sorted_temp_a = dataFrame.sort_values(by='TEMP')
print(sorted_temp_a)

# sort values in descending order
sorted_temp_d = dataFrame.sort_values(by='TEMP', ascending=False)
print(sorted_temp_d)
```
